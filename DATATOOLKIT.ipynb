Q1. What is NumPy, and why is it widely used in Python?

ANS - NumPy is a powerful Python library used for numerical and scientific computing. It provides efficient support for multidimensional arrays and mathematical functions that operate on these arrays. NumPy is widely used because it offers high performance, low memory consumption, and fast execution through optimized C-based implementations. It also supports vectorized operations, which remove the need for loops. For example, NumPy is commonly used to perform matrix multiplication in machine learning applications.

Q2. How does broadcasting work in NumPy?

ANS - Broadcasting in NumPy allows arithmetic operations between arrays of different shapes by automatically expanding the smaller array to match the larger one. This eliminates the need for explicit loops and improves computational efficiency. Broadcasting follows specific rules related to array dimensions and compatibility. It simplifies coding and makes array operations more intuitive. For example, adding a scalar value to all elements of an array uses broadcasting.

Q3. What is a Pandas DataFrame?

ANS - A Pandas DataFrame is a two-dimensional, labeled data structure used to store data in rows and columns. It can handle different data types within the same table and provides powerful data manipulation features. DataFrames are easy to analyze, filter, and modify, making them ideal for data analysis tasks. They are similar to Excel spreadsheets or database tables. For example, a DataFrame can store student records with names, marks, and grades.

Q4. Explain the use of the groupby() method in Pandas.

ANS-The groupby() method in Pandas is used to group data based on one or more columns so that aggregate operations can be performed. It enables split-apply-combine operations, allowing users to compute sums, averages, or counts for each group. This method is widely used in data analysis for summarizing large datasets. It improves data organization and insights. For example, groupby() can be used to calculate average salary for each department.

Q5. Why is Seaborn preferred for statistical visualizations?

Built on top of Matplotlib, it simplifies the creation of complex statistical graphs such as box plots, violin plots, and heatmaps. Seaborn integrates well with Pandas DataFrames and automatically handles data aggregation. It also offers better default aesthetics.

Q6. What are the differences between NumPy arrays and Python lists?

ANS- NumPy arrays differ from Python lists in terms of performance, memory usage, and functionality. NumPy arrays store homogeneous data and support vectorized operations, making them faster and more efficient. Python lists can store heterogeneous data but are slower for numerical computations. NumPy arrays also consume less memory and support advanced mathematical operations. For example, element-wise addition is faster in NumPy arrays than in lists.

Q7. What is a heatmap, and when should it be used?

It is commonly used to identify patterns, trends, and correlations in large datasets. Heatmaps make complex numerical data easy to interpret visually. They are especially useful in statistical analysis. For example, a correlation matrix is often visualized using a heatmap.

Q8. What does the term “vectorized operation” mean in NumPy?

ANS-A vectorized operation in NumPy refers to performing computations on entire arrays without using explicit loops. These operations are executed at a low level, making them faster and more efficient. Vectorization improves code readability and performance. It is one of the main advantages of using NumPy. For example, multiplying all elements of an array by a constant is a vectorized operation.

Q1- How do you create a 2D NumPy array and calculate the sum of each row?

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])
row_sum = arr.sum(axis=1)
print(row_sum)
```

Q2- Write a Pandas script to find the mean of a specific column in a DataFrame?

```python
import pandas as pd

df = pd.DataFrame({
    'Marks': [70, 80, 90, 85]
})
mean_value = df['Marks'].mean()
print(mean_value)
```

Q3 - Create a scatter plot using Matplotlib?

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

Q4 - How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

corr = df.corr()
sns.heatmap(corr, annot=True)
```

Q13- Create a 3D scatter plot using Plotly?

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})

fig = px.scatter_3d(df, x='x', y='y', z='z')
fig.show()
```
