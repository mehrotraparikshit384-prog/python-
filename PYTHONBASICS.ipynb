Question - 13  What is the difference between for and while loops

Answer - The for loop in Python is used when you know in advance how many times you want to repeat a block of code. It works with sequences like lists, strings, and the range() function, and it automatically goes through each item in the sequence. On the other hand, a while loop is used when you want a block of code to run repeatedly until a certain condition becomes false, meaning you do not know the exact number of repetitions beforehand. In simple terms, a for loop is ideal for fixed or countable repetitions, while a while loop is better when the loop depends on a condition that changes during the program's execution

Question - 5  Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
for each.

```python
#ANSWER
# list (mutable)
lst = [1, 2, 3]
print("Before list change:", lst)
lst[0] = 100
print("After list change:", lst)

# tuple (immutable)
tpl = (1, 2, 3)
print("Tuple before attempted change:", tpl)
try:
    tpl[0] = 100
except TypeError as e:
    print("Error trying to change tuple element:", e)
```

Question 6     Write a function to demonstrate the behavior of mutable and immutable arguments

```python
#ANSWER
def modify_list(a_list):
    a_list.append(99)

def modify_int(a_int):
    a_int += 10
    return a_int

lst = [1, 2]
num = 5

print("Before modify_list:", lst)
modify_list(lst)
print("After modify_list (list changed):", lst)

print("Before modify_int:", num)
num2 = modify_int(num)
print("After modify_int (original unchanged):", num)
print("Returned value from modify_int:", num2)

```

QUESTION - 9 . Write a Python program to convert user input from string to integer, float, and boolean types.

```python

s = input("Enter something: ")

# to int (safe conversion with try/except)
try:
    i = int(s)
    print("As int:", i)
except ValueError:
    print("Cannot convert to int.")

# to float
try:
    f = float(s)
    print("As float:", f)
except ValueError:
    print("Cannot convert to float.")

# to bool (note: bool("any non-empty string") is True; to parse "True"/"False" use lower-check)
b = s.lower() in ("true", "1", "yes", "y", "t")
print("Interpreted boolean (True/False/1/yes accepted):", b)
```

QUESTION 10 Write code to demonstrate type casting with list elements


```python
#ANSWER
str_list = ["1", "2", "3", "10"]
print("Original (strings):", str_list)

# convert to ints
int_list = [int(x) for x in str_list]
print("Converted to ints:", int_list)

# convert back to strings
str_again = [str(x) for x in int_list]
print("Back to strings:", str_again)
```

QUESTION 14 . Write a program to reverse a string using a while loop

```python
#ANSWER
s = input("Enter a string to reverse: ")
i = len(s) - 1
rev = ""
while i >= 0:
    rev += s[i]
    i -= 1
print("Reversed string:", rev)
```

QUESTION 15 . Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
#ANSWER
n = int(input("Enter a non-negative integer for factorial: "))
if n < 0:
    print("Factorial not defined for negative numbers.")
else:
    fact = 1
    i = 1
    while i <= n:
        fact *= i
        i += 1
    print(f"Factorial of {n} is {fact}")
```
