Question 1: What is hypothesis testing in statistics?


ANSWER - Hypothesis testing is a statistical procedure used to determine whether there is sufficient evidence from sample data to make an inference about a population parameter. It helps in deciding whether a given assumption about a population is true or not.




Types of Hypotheses -

Null Hypothesis (H₀)
It states that there is no significant difference or effect.
Example: The mean accuracy of an ML model is 90%.

Alternative Hypothesis (H₁)
It states that there is a significant difference or effect.
Example: The mean accuracy of the ML model is not 90%.



Steps in Hypothesis Testing -

Formulate H₀ and H₁

Select level of significance (α), usually 0.05

Choose appropriate statistical test (z-test, t-test, chi-square test, ANOVA)

Compute test statistic

Calculate p-value

Decision making:

If p ≤ α → reject H₀

If p > α → accept H₀

Example

A company claims a machine learning model has 95% accuracy. A sample test gives 92% accuracy with p-value 0.03. Since 0.03 < 0.05, the null hypothesis is rejected, meaning the claim is not valid.

Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?

ANSWER  - Null Hypothesis (H₀)

The null hypothesis is a statement that assumes there is no effect, no difference, or no relationship between variables. It represents the default or initial assumption that any observed difference is due to random chance.

Example:
H₀: The average accuracy of an ML model is 90%.

Alternative Hypothesis (H₁ or Hₐ)

The alternative hypothesis is a statement that contradicts the null hypothesis. It claims that there is a significant effect, difference, or relationship between variables.

Example:
H₁: The average accuracy of the ML model is not 90%.

Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.


ANSWER - The significance level, denoted by α (alpha), is the probability of rejecting the null hypothesis when it is actually true. It represents the maximum risk a researcher is willing to take of making a Type I error.

Commonly used values of α are 0.05 (5%), 0.01 (1%), and 0.10 (10%).

Role of Significance Level in Hypothesis Testing

Sets the Decision Boundary
The significance level determines the cutoff point for deciding whether the test result is statistically significant.

Comparison with p-value

If p-value ≤ α → Reject the null hypothesis

If p-value > α → Fail to reject the null hypothesis

Controls Type I Error
By fixing α, researchers control the probability of incorrectly rejecting a true null hypothesis.

Determines Confidence in Results
A lower α (e.g., 0.01) requires stronger evidence to reject H₀, giving more reliable results

Question 4: What are Type I and Type II errors? Give examples of each.


ANSWER - Type I Error (α Error)

A Type I error occurs when the null hypothesis is rejected even though it is true.
It is also known as a false positive.

Probability of Type I error = α (significance level)

Example:
H₀: A machine is working correctly.
Decision: The test rejects H₀ and concludes the machine is faulty, but in reality it is working fine.

Type II Error (β Error)

A Type II error occurs when the null hypothesis is not rejected even though it is false.
It is also known as a false negative.

Probability of Type II error = β

Example:
H₀: A patient does not have a disease.
Decision: The test fails to reject H₀, but the patient actually has the disease.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.

ANSWER -

Z-test-

1-A Z-test is used when:

2-Sample size is large (n ≥ 30)

3-Population standard deviation (σ) is known

4- Data is normally distributed (or sample is large enough by CLT)


T-test-

1- A T-test is used when:

2-Sample size is small (n < 30)

3-Population standard deviation is unknown

4- Sample standard deviation (s) is used instead of σ

Question 6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.


Hint: Generate random number using random function.


```python
import random
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
trials = 10000

# Generate binomial distribution using random function
data = []

for _ in range(trials):
    success = 0
    for _ in range(n):
        if random.random() < p:
            success += 1
    data.append(success)

# Plot histogram
plt.hist(data, bins=range(0, n+2))
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
n = 100
mu = 50
sigma = 10

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Step 3: Compute 95% confidence interval
z = 1.96  # for 95% confidence level
margin_of_error = z * (sample_std / sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot the data
plt.figure()
plt.hist(data, bins=15)
plt.axvline(sample_mean)
plt.axvline(ci_lower, linestyle='--')
plt.axvline(ci_upper, linestyle='--')
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores

# Example dataset
data = np.array([45, 50, 52, 48, 47, 55, 60, 53, 49, 51])

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Z-scores:", z_scores)

# Plot histogram of standardized data
plt.figure()
plt.hist(z_scores, bins=8)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
