Q1- What is the difference between descriptive statistics and inferential
statistics? Explain with examples?

ANS- Descriptive statistics and inferential statistics differ mainly in their purpose and scope of analysis. Descriptive statistics are used to organize, summarize, and present data in a meaningful way, without drawing conclusions beyond the given data. They describe what the data shows using measures such as mean, median, mode, range, variance, standard deviation, tables, graphs, and charts. For example, calculating the average marks of a class and displaying them using a bar chart is descriptive statistics, as it only summarizes the existing data. In contrast, inferential statistics are used to draw conclusions, make predictions, or generalize results about a population based on a sample of data. It involves probability, hypothesis testing, confidence intervals, correlation, and regression.


For example, using the marks of a sample of students to predict the average performance of all students in a university is inferential statistics. Thus, descriptive statistics describe data, while inferential statistics help in decision-making and prediction beyond the observed data.

Q2- : What is sampling in statistics? Explain the differences between random
and stratified sampling?

ANS-   Sampling in statistics is the process of selecting a subset (sample) of individuals, items, or observations from a larger group (population) to study and analyze, so that conclusions about the population can be drawn efficiently and economically. Instead of collecting data from every member of the population, sampling saves time, cost, and effort while still providing reliable results when done properly.

For example, surveying 200 voters out of a city of 20,000 voters to understand voting preferences is sampling.

Random sampling and stratified sampling differ in how the sample is selected. In random sampling, every member of the population has an equal and independent chance of being selected, and selection is made purely by chance, such as using a lottery method or random number table. For example, randomly choosing 50 students from a college register is random sampling. In stratified sampling, the population is first divided into homogeneous subgroups called strata (such as age groups, gender, or departments), and then samples are randomly selected from each stratum in proportion to their size.


For example, selecting students separately from first-year, second-year, and third-year classes ensures proper representation. Thus, random sampling ensures simplicity and fairness, while stratified sampling improves accuracy by representing all important subgroups of the population.

Q3- Define mean, median, and mode. Explain why these measures of central
tendency are important?

ANS- Mean, median, and mode are measures of central tendency used to represent a large set of data by a single typical value. The mean is defined as the arithmetic average of all observations and is calculated by dividing the sum of all values by the number of observations; for example, the mean of 2, 4, and 6 is (2+4+6)/3 = 4. The median is the middle value of a data set when the observations are arranged in ascending or descending order; for example, for the data 3, 5, 7, 9, and 11, the median is 7. The mode is the value that occurs most frequently in a data set; for example, in the data 2, 4, 4, 6, and 8, the mode is 4.

These measures of central tendency are important because they help in summarizing and understanding large volumes of data in a simple and meaningful way. The mean is widely used in calculations and comparisons, the median is useful when data contains extreme values as it is not affected by outliers, and the mode is helpful in identifying the most common or popular value, especially in categorical data. Together, they provide a clear picture of the data’s central behavior and support effective analysis and decision-making.

Q4- Explain skewness and kurtosis. What does a positive skew imply about the
data?

ANS- Skewness and kurtosis are statistical measures used to describe the shape of a data distribution. Skewness refers to the degree of asymmetry of a distribution around its mean. A distribution is said to be symmetrical if the left and right sides of the curve are mirror images of each other. When the distribution has a longer tail on the right side, it is called positive skewness, and when it has a longer tail on the left side, it is called negative skewness. For example, income distribution is usually positively skewed, where most people earn moderate incomes but a few earn very high incomes.

Kurtosis refers to the degree of peakedness or flatness of a distribution compared to a normal distribution. A distribution may be leptokurtic (highly peaked), mesokurtic (normal peak), or platykurtic (flatter than normal). For example, a distribution with many values close to the mean and heavy tails is leptokurtic.

A positive skew implies that most of the data values are concentrated on the lower side of the distribution, with a few extremely high values stretching the tail to the right. In such cases, the mean is greater than the median, and the median is greater than the mode (Mean > Median > Mode).

Q5 -  Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
(Include your Python code and output in the code box below.)?


```python
# Python program to compute mean, median, and mode

import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.multimode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

Q6- Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

```python
# Python program to compute covariance and correlation coefficient

import statistics

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

mean_x = statistics.mean(list_x)
mean_y = statistics.mean(list_y)

# Compute sample covariance
covariance = sum((x - mean_x) * (y - mean_y) for x, y in zip(list_x, list_y)) / (len(list_x) - 1)

# Compute correlation coefficient
std_x = statistics.stdev(list_x)
std_y = statistics.stdev(list_y)
correlation = covariance / (std_x * std_y)

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

Q8 - You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.


● Explain how you would use covariance and correlation to explore this
relationship.


● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

ANS - Using Covariance and Correlation to Analyze the Relationship

To explore the relationship between advertising spend and daily sales, I would use covariance and correlation as statistical tools. Covariance helps determine the direction of the relationship between the two variables. If the covariance is positive, it indicates that higher advertising spend is associated with higher sales; if negative, higher spend is associated with lower sales. However, covariance does not indicate the strength of the relationship. To measure both the strength and direction, correlation is used. The correlation coefficient ranges from –1 to +1, where values close to +1 indicate a strong positive relationship, values close to –1 indicate a strong negative relationship, and values near 0 indicate little or no linear relationship. Together, these measures help the marketing team understand whether increasing advertising spend is likely to increase daily sales.

```python
import statistics

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute correlation coefficient
correlation = statistics.correlation(advertising_spend, daily_sales)

print("Correlation Coefficient:", correlation)
```

Q9 - Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.

● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.

● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

ANS - Summary Statistics and Visualizations to Use

 To understand the distribution of customer satisfaction scores before launching a new product, I would first use summary statistics such as the mean to understand the average satisfaction level, the median to find the central score unaffected by extreme values, and the standard deviation to measure how spread out the responses are. These statistics give a quick numerical summary of customer sentiment. Along with this, I would use a histogram as a visualization tool to observe the overall shape of the distribution, identify whether scores are clustered around certain values, and check for skewness or gaps. This combination helps in assessing whether customer satisfaction is generally high, moderate, or varied.

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.figure()
plt.hist(survey_scores, bins=7)
plt.title("Histogram of Customer Satisfaction Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")
plt.show()
```
