Q2 -Difference between mutable and immutable data types ?

Ans -
      
      Mutable-

           Can be changed after creation.

           Examples: list, dictionary, set

      Immutable -

          Cannot be changed once created.

           Examples: tuple, string, int

Q3 -  What are the main differences between lists and tuples in Python?


Ans -   
        
        Lists are mutable; tuples are immutable.

        Lists use []; tuples use ().

        Lists are slower; tuples are faster in access.

        Tuples are used for fixed data; lists for dynamic data.
